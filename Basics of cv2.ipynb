```python
#How to put text on image.....
import cv2
cap=cv2.VideoCapture(0)
ret,fram=cap.read()
print(ret)
font=cv2.FONT_HERSHEY_COMPLEX
fram1=cv2.putText(fram,"hello",(200,200),font,4,(255,110,10),2,cv2.LINE_AA)
cv2.imshow("komal",fram1)
cv2.waitKey(0)
cv2.destroyAllWindows()
cap.release()
```

```python
#How to show video cammera.....
import cv2
cap=cv2.VideoCapture(0)
k=0
while(k!=27):
    ret,fram=cap.read()
    print(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    cv2.imshow("komal",fram)
    k=cv2.waitKey(1)
cv2.destroyAllWindows()
cap.release()
```

```python
#How to get frame width and height cammera.....
import cv2
cap=cv2.VideoCapture(0)
k=0
while(k!=27):
    ret,fram=cap.read()
    print(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    print(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    cv2.imshow("komal",fram)
    k=cv2.waitKey(1)
cv2.destroyAllWindows()
cap.release()
```

```python
#How to convert color to gray image cammera.....
import cv2
cap=cv2.VideoCapture(0)
k=0

while(k!=27):
    ret,fram=cap.read()
    x=cap.get(cv2.CAP_PROP_FRAME_WIDTH)
    y=cap.get(cv2.CAP_PROP_FRAME_HEIGHT)
    font=cv2.FONT_HERSHEY_COMPLEX
    fram=cv2.putText(fram,'Width='+str(x)+',',(0,20),font,0.5,(255,255,255),1,cv2.LINE_AA)
    fram=cv2.putText(fram,'Height='+str(y),(110,20),font,0.5,(255,255,255),1,cv2.LINE_AA)
    fram=cv2.cvtColor(fram,cv2.COLOR_BGR2GRAY)
    cv2.imshow("komal",fram)
    k=cv2.waitKey(1)
cv2.destroyAllWindows()
cap.release()
```

```python
#How to set frame property.....
import cv2
cap=cv2.VideoCapture(0)
cap.set(3,1000)
cap.set(4,1000)
k=0

while(k!=27):
    ret,fram=cap.read()
   
    
    font=cv2.FONT_HERSHEY_COMPLEX
    fram=cv2.putText(fram,'Width='+str(cap.get(3))+',',(0,20),font,0.5,(255,255,255),1,cv2.LINE_AA)
    fram=cv2.putText(fram,'Height='+str(cap.get(4)),(110,20),font,0.5,(255,255,255),1,cv2.LINE_AA)
    cv2.imshow("komal",fram)
    k=cv2.waitKey(1)
cv2.destroyAllWindows()
cap.release()
```

```python
#How to display date and time  property.....
import cv2
import datetime
cap=cv2.VideoCapture(0)
cap.set(3,1000)
cap.set(4,1000)
k=0


while(k!=27):
    ret,fram=cap.read()
    font=cv2.FONT_HERSHEY_COMPLEX
    date=str(datetime.datetime.now())
    fram=cv2.putText(fram,'Width='+str(cap.get(3))+',',(0,20),font,0.5,(255,255,255),1,cv2.LINE_AA)
    fram=cv2.putText(fram,'Height='+str(cap.get(4)),(120,20),font,0.5,(255,255,255),1,cv2.LINE_AA)
    fram=cv2.putText(fram,'Date='+date,(0,40),font,0.5,(255,255,255),1,cv2.LINE_AA)
    cv2.imshow("komal",fram)
    k=cv2.waitKey(1)
cv2.destroyAllWindows()
cap.release()
```
